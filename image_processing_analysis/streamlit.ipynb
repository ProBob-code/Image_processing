```python
# convert the data types of the columns
new['product_id'] = new['product_id'].astype(int)
new['product_url'] = new['product_url'].astype(str)
new['keywords'] = new['keywords'].astype(str)
new['image_name'] = new['image_name'].astype(str)
new['height'] = new['height'].astype(int)
new['width'] = new['width'].astype(int)
new['size'] = new['size'].astype(float)
new['image_format'] = new['image_format'].astype(str)
new['image_mode'] = new['image_mode'].astype(str)
new['description'] = new['description'].astype(str)
new['author'] = new['author'].astype(str)
new['copyright'] = new['copyright'].astype(str)
new['location'] = new['location'].astype(str)
new['image_shape'] = new['image_shape'].astype(str)
new['megapixels'] = new['megapixels'].astype(float)
new['ppi'] = new['ppi'].astype(int)
new['exif_dict'] = new['exif_dict'].astype(str)
new['description'] = new['description'].astype(str)
new['author'] = new['author'].astype(str)
new['copyright'] = new['copyright'].astype(str)
new['location'] = new['location'].astype(str)
new['image_shape'] = new['image_shape'].astype(str)
new['megapixels'] = new['megapixels'].astype(float)
new['ppi'] = new['ppi'].astype(int)
new['exif_dict'] = new['exif_dict'].astype(str)
new['red'] = new['red'].astype(float)
new['green'] = new['green'].astype(float)
new['blue'] = new['blue'].astype(float)
new['laplacian_variance_blur'] = new['laplacian_variance_blur'].astype(float)
new['fourier_transform_blur'] = new['fourier_transform_blur'].astype(float)
new['gradient_magnitude_blur'] = new['gradient_magnitude_blur'].astype(float)
new['image_metric'] = new['image_metric'].astype(float)
new['duplicate'] = new['duplicate'].astype(int)


df = new

P_h = np.average(df.height)
P_w = np.average(df.width)
P_s = np.average(df.size)
P_ppi= np.average(df.ppi)
P_mp = np.average(df.megapixels)
P_lvb = np.average(df.laplacian_variance_blur)
P_ftb = np.average(df.fourier_transform_blur)
P_gmb = np.average(df.gradient_magnitude_blur)

column_names = ['product_id','image_blur_flag','image_quality_flag']
df3 = pd.DataFrame(columns=column_names)

blur = []
for i in range(len(df)):
    if df['height'][i] < P_h:
        Q_h = int(0)
    else:
        Q_h = int(1)

    if df['width'][i] < P_w:
        Q_w = int(0)
    else:
        Q_w = int(1)

    if df['size'][i] < P_s:
        Q_s = int(0)
    else:
        Q_s = int(1)

    if df['ppi'][i] < P_ppi:
        Q_ppi = int(0)
    else:
        Q_ppi = int(1)

    if df['megapixels'][i] < P_mp:
        Q_mp = int(0)
    else:
        Q_mp = int(1)

    if df['laplacian_variance_blur'][i] < P_lvb:
        Q_lvb = int(0)
    else:
        Q_lvb = int(1)
    
    if df['fourier_transform_blur'][i] < P_ftb:
        Q_ftb = int(0)
    else:
        Q_ftb = int(1)

    if df['gradient_magnitude_blur'][i] < P_gmb:
        Q_gmb = int(0)
    else:
        Q_gmb = int(1)

    if (Q_lvb + Q_ftb + Q_gmb) / 3 < 1:
        blur.append(int(0))
    else:
        blur.append(int(1))


    if (Q_w + Q_s + Q_h + Q_ppi + Q_mp + Q_lvb + Q_ftb + Q_gmb) / 8 < 0.62:
        df3.loc[i] = [df['product_id'][i],blur[i],int(0)]
    else:
        df3.loc[i] =[df['product_id'][i],blur[i],int(1)]



df3['image_quality_flag'].value_counts()
df = pd.concat([df,df3['image_quality_flag']],axis=1)

# df['duplicate'] = 0
duplicate = df.duplicated(subset=['image_metric'], keep=False)
df.loc[duplicate, 'duplicate'] = 1

# duplicate_values = df[df.duplicated(subset=['image_metric'], keep=False)].sort_values(by=['image_metric'])



def barplot_data(data, variables):
    for variable in variables:
        # Barplot
        plt.figure(figsize=(15, 10))
        sns.barplot(x=variable, y='count', data=data.groupby(variable).size().reset_index(name='count'))
        plt.title(f'Barplot of {variable}')
        plt.show()

def hist_data(data, variables):       
    for variable in variables:
        # Histogram
        plt.figure(figsize=(15, 10))
        sns.histplot(data[variable], kde=True)
        plt.title(f'Histogram of {variable}')
        plt.show()

def pie_data(data, variables):
    for variable in variables: 
        # Pie Chart
        plt.style.use('default')
        plt.figure(figsize=(15, 10))
        plt.pie(data.groupby(variable).size(), labels=(data[variable].unique()),explode=(0.1, 0.1), autopct='%1.2f%%',shadow=True)
        plt.legend()
        plt.title(f'Pie Chart of {variable}')
        plt.show()

visual_data = df.drop(['image_shape'], axis=1)

# Load data
data = visual_data



# # Specify variables to plot
# variables1 = ['duplicate',
#        'image_quality_flag']

# # Specify variables to plot
# variables2 = ['height','width','size','ppi','red','blue','green']

# # variables3 = ['Duplicate',
# #        'Image_Quality_flag']

# # Generate plots
# # barplot_data(data, variables1)
# pie_data(data,variables1)
# hist_data(data,variables2)
```
